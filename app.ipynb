# 📊 Global Education Statistics Dashboard

## 🔍 Project Overview

This project is an interactive data visualization dashboard built using **Dash** and **Plotly**.
It provides insights into global education statistics such as literacy rates, unemployment, birth rates, and gender gaps across different countries.

The dashboard allows users to:

* Filter data by country
* Select number of top countries (Top N)
* Switch between different metrics
* Explore multiple chart types (bar, scatter, bubble, histogram, etc.)

The goal of this project is to make education-related data **interactive, clear, and easy to analyze**.

---

## 📁 Dataset Description

The dataset used in this project contains global education and demographic indicators.

**Main Features:**

* `country` → Country name
* `literacy_rate` → Overall literacy rate (%)
* `literacy_male` → Male literacy rate
* `literacy_female` → Female literacy rate
* `literacy_gap` → Difference between male & female literacy
* `unemployment_rate` → Unemployment percentage
* `birth_rate` → Birth rate
* `primary_enrollment` → Primary education enrollment
* `tertiary_enrollment` → Higher education enrollment
* `g2_reading` → Reading performance score

📌 The dataset is loaded from a local CSV file in the project.

# 1) COLUMN CHART

````python
# ==============================
# COLUMN CHART
# ==============================
@app.callback(
    Output("column-chart", "figure"),
    Input("country-dropdown", "value"),
    Input("top-n-slider", "value"),
)

def update_column_chart(selected_countries, top_n):

    metric = "literacy_gap"

    filtered_df = df.copy()

    if selected_countries:
        filtered_df = filtered_df[
            filtered_df["country"].isin(selected_countries)
        ]

    if filtered_df.empty:
        return px.bar(title="No Data Available")

    # SORT
    filtered_df = filtered_df.sort_values(
        by=metric, ascending=False
    ).head(top_n)

    # ======================
    # COLORS (light + highlight)
    # ======================
    colors = ["#AED6F1"] * len(filtered_df)

    if len(colors) > 0:
        colors[0] = "#58D68D"

    # ======================
    # PLOT
    # ======================
    fig = px.bar(
        filtered_df,
        x="country",
        y=metric,
        text=metric
    )

    # ======================
    # STYLE
    # ======================
    fig.update_traces(
        marker_color=colors,
        texttemplate='%{text:.1f}',
        textposition='outside'
    )

    fig.update_layout(
    title=dict(
        text=f"Top {top_n} Countries with the Highest Literacy Gap <br> Between Males and Females",
        x=0.5
    ),

    plot_bgcolor="white",
    paper_bgcolor="white",

    xaxis=dict(
        title="Country",
        showgrid=True,
        gridcolor="lightgray",
        showline=True,
        linecolor="black",
        mirror=True,
        tickangle=-30
    ),

    yaxis=dict(
        title="Literacy Gap",
        showgrid=True,
        gridcolor="lightgray",
        showline=True,
        linecolor="black",
        mirror=True,
        rangemode="tozero"
    ),

    legend=dict(
        x=0.98,
        y=0.98,
        xanchor="right",
        yanchor="top",
        bgcolor="white",
        bordercolor="black",
        borderwidth=1,
        title="Category"
    ),

    margin=dict(l=40, r=40, t=60, b=40),

    shapes=[
        dict(
            type="rect",
            xref="paper",
            yref="paper",
            x0=0, y0=0,
            x1=1, y1=1,
            line=dict(color="black", width=2),
            fillcolor="rgba(0,0,0,0)"
        )
    ]
)
    
    fig.add_scatter(
        x=[None], y=[None],
        mode='markers',
        marker=dict(color="#D6EAF8"),
        name="Other Countries"
    )
    
    fig.add_scatter(
        x=[None], y=[None],
        mode='markers',
        marker=dict(color="#82E0AA"),
        name="Highest Value"
    )
    return fig
````

# 2) BAR CHART

````python
# ============================== 
# BAR CHART 
# ==============================

@app.callback(
    Output("bar-chart", "figure"),
    Input("country-dropdown", "value"),
    Input("top-n-slider", "value"),
    Input("metric-radio", "value")   
)
def update_bar_chart(selected_countries, top_n, metric):

    filtered_df = df.copy()

    # ======================
    # FILTER
    # ======================
    if selected_countries:
        filtered_df = filtered_df[
            filtered_df["country"].isin(selected_countries)
        ]

    filtered_df = filtered_df[filtered_df[metric] > 0]

    if filtered_df.empty:
        return px.bar(title="No Data Available")

    # ======================
    # SORT
    # ======================
    filtered_df = filtered_df.sort_values(
        by=metric, ascending=True
    ).head(top_n)

    # ======================
    # COLORS
    # ======================
    colors = ["#D6EAF8"] * len(filtered_df)

    if len(colors) > 0:
        colors[0] = "#82E0AA"

    # ======================
    # PLOT
    # ======================
    fig = px.bar(
        filtered_df,
        y="country",
        x=metric,                 #  dynamic
        text=metric,              #  dynamic
        orientation="h"
    )

    # ======================
    # STYLE
    # ======================
    fig.update_traces(
        marker_color=colors,
        texttemplate='%{text:.1f}',
        textposition='outside'
    )

    fig.update_layout(
    title=dict(
        text=f"Top {top_n} Countries with the Lowest {metric.replace('_',' ').title()}",
        x=0.5
    ),

    plot_bgcolor="white",
    paper_bgcolor="white",

    xaxis=dict(
        title=metric.replace("_", " ").title(),
        showgrid=True,
        gridcolor="lightgray",
        showline=True,
        linecolor="black",
        mirror=True,
        rangemode="tozero"
    ),

    yaxis=dict(
        title="Country",
        showgrid=True,
        gridcolor="lightgray",
        showline=True,
        linecolor="black",
        mirror=True
    ),

    legend=dict(
        x=0.98,
        y=0.98,
        xanchor="right",
        yanchor="top",
        bgcolor="white",
        bordercolor="black",
        borderwidth=1
    ),

    margin=dict(l=100, r=20, t=60, b=50),

    shapes=[
        dict(
            type="rect",
            xref="paper",
            yref="paper",
            x0=0, y0=0,
            x1=1, y1=1,
            line=dict(color="black", width=2),
            fillcolor="rgba(0,0,0,0)"
        )
    ]
)

    # ======================
    # LEGEND
    # ======================
    fig.add_scatter(
        x=[None], y=[None],
        mode='markers',
        marker=dict(color="#D6EAF8"),
        name="Other Countries"
    )

    fig.add_scatter(
        x=[None], y=[None],
        mode='markers',
        marker=dict(color="#82E0AA"),
        name="Lowest Value"
    )

    return fig
````

# 7) SCATTER CHART

````python
import plotly.graph_objects as go

def create_scatter_figure(df):

    x = 'g2_reading'
    y = 'literacy_rate'

    # 🚨 DROP NA (مهم جدًا)
    df = df.dropna(subset=[x, y])

    # ======================
    # OUTLIER DETECTION (IQR)
    # ======================
    Q1_x, Q3_x = df[x].quantile(0.25), df[x].quantile(0.75)
    Q1_y, Q3_y = df[y].quantile(0.25), df[y].quantile(0.75)

    IQR_x = Q3_x - Q1_x
    IQR_y = Q3_y - Q1_y

    outliers = df[
        (df[x] > Q3_x + 1.5 * IQR_x) |
        (df[x] < Q1_x - 1.5 * IQR_x) |
        (df[y] > Q3_y + 1.5 * IQR_y) |
        (df[y] < Q1_y - 1.5 * IQR_y)
    ]

    normal = df.drop(outliers.index)
    # ======================
    # FIGURE
    # ======================
    fig = go.Figure()

    # NORMAL POINTS
    fig.add_trace(go.Scatter(
        x=normal[x],
        y=normal[y],
        mode='markers',
        name='Standard Performance',
        text=normal['country'],
        marker=dict(
            color='lightblue',
            size=10,
            line=dict(color='black', width=1),
            opacity=0.6
        ),
        hovertemplate=
        "<b>%{text}</b><br>" +
        "G2 Reading: %{x}<br>" +
        "Literacy: %{y}%<extra></extra>"
    ))

    # OUTLIERS
    fig.add_trace(go.Scatter(
        x=outliers[x],
        y=outliers[y],
        mode='markers+text',
        name='Performance Outlier',
        text=outliers['country'],
        textposition='top center',
        marker=dict(
            color='#f4c27a',
            size=12,
            line=dict(color='black', width=1),
            opacity=0.9
        ),
        hovertemplate=
        "<b>%{text}</b><br>" +
        "G2 Reading: %{x}<br>" +
        "Literacy: %{y}%<extra></extra>"
    ))

    # ======================
    # LAYOUT
    # ======================
    fig.update_layout(
        title=dict(
            text="Relationship Between Reading Performance (G2 Score) and Literacy Rate <br> with Outlier Detection",
            x=0.5
        ),
        plot_bgcolor="white",
        paper_bgcolor="white",

        xaxis=dict(
            title="G2 Reading Score",
            showgrid=True,
            gridcolor="lightgray",
            showline=True,
            linecolor="black",
            mirror=True
        ),

        yaxis=dict(
            title="Literacy Rate (%)",
            showgrid=True,
            gridcolor="lightgray",
            showline=True,
            linecolor="black",
            mirror=True,
            range=[0, df[y].max() * 1.05]  
        ),

        legend=dict(
            x=0.98,
            y=0.98,
            xanchor="right",
            yanchor="top",
            bgcolor="white",
            bordercolor="black",
            borderwidth=1
        )
    )

    return fig
````

# 8) BUBBLE CHART

````python
import plotly.graph_objects as go
import numpy as np

def create_bubble_figure(df):

    x = 'birth_rate'
    y = 'literacy_rate'
    size_col = 'literacy_gap'

    # ======================
    # CLEAN DATA
    # ======================
    df_clean = df.copy()
    df_clean = df_clean.dropna(subset=[x, y, size_col]).reset_index(drop=True)

    if df_clean.empty:
        return go.Figure().update_layout(title="No Data Available")

    # ======================
    # SIZE (safe scaling 🔥)
    # ======================
    size = df_clean[size_col].abs()
    size = size.fillna(0)

    # cap outliers
    size = size.clip(upper=size.quantile(0.9))

    # 🔥 FIX division by zero
    if size.max() == size.min():
        size_scaled = np.full(len(size), 10)
    else:
        size_scaled = (size - size.min()) / (size.max() - size.min()) * 20 + 5

    # ======================
    # OUTLIERS (IQR)
    # ======================
    Q1_x, Q3_x = df_clean[x].quantile(0.25), df_clean[x].quantile(0.75)
    Q1_y, Q3_y = df_clean[y].quantile(0.25), df_clean[y].quantile(0.75)

    IQR_x = Q3_x - Q1_x
    IQR_y = Q3_y - Q1_y

    outliers = df_clean[
        (df_clean[x] > Q3_x + 1.5 * IQR_x) |
        (df_clean[x] < Q1_x - 1.5 * IQR_x) |
        (df_clean[y] > Q3_y + 1.5 * IQR_y) |
        (df_clean[y] < Q1_y - 1.5 * IQR_y)
    ]

    normal = df_clean.drop(outliers.index)

    # ======================
    # FIGURE
    # ======================
    fig = go.Figure()

    # NORMAL POINTS
    fig.add_trace(go.Scatter(
        x=normal[x],
        y=normal[y],
        mode='markers',
        name='Standard Performance',
        marker=dict(
            size=size_scaled[normal.index],
            color='lightblue',
            opacity=0.5,
            line=dict(color='black', width=1)
        ),
        hovertemplate=
            "<b>%{text}</b><br>" +
            "Birth Rate: %{x}<br>" +
            "Literacy Rate: %{y}%<extra></extra>",
        text=normal['country']
    ))

    # OUTLIERS
    fig.add_trace(go.Scatter(
        x=outliers[x],
        y=outliers[y],
        mode='markers+text',
        name='Performance Outlier',
        text=outliers['country'],
        textposition='top center',
        textfont=dict(size=10, color='black'),
        marker=dict(
            size=size_scaled[outliers.index],
            color='#f4c27a',
            opacity=0.8,
            line=dict(color='black', width=1)
        ),
        hovertemplate=
            "<b>%{text}</b><br>" +
            "Birth Rate: %{x}<br>" +
            "Literacy Rate: %{y}%<extra></extra>"
    ))

    # ======================
    # LAYOUT
    # ======================
    fig.update_layout(
        title=dict(
            text="Relationship Between Birth Rate and Literacy Rate <br> with Gender Gap Represented by Bubble Size",
            x=0.5
        ),

        plot_bgcolor="white",
        paper_bgcolor="white",

        xaxis=dict(
            title="Birth Rate",
            showgrid=True,
            gridcolor="lightgray",
            showline=True,
            linecolor="black",
            linewidth=2,
            mirror=True
        ),

        yaxis=dict(
            title="Literacy Rate (%)",
            showgrid=True,
            gridcolor="lightgray",
            showline=True,
            linecolor="black",
            linewidth=2,
            mirror=True,
            range=[0, df_clean[y].max() * 1.05]
        ),

        legend=dict(
            x=0.98,
            y=0.98,
            xanchor="right",
            yanchor="top",
            bgcolor="white",
            bordercolor="black",
            borderwidth=1
        )
    )

    return fig
````

````python
import plotly.graph_objects as go

def create_line_figure(df, metric, top_n=20):

    # ======================
    # TOP N 
    # ======================
    d = df.sort_values(by=metric, ascending=False).head(top_n)

    d = d.sort_values(by=metric).reset_index(drop=True)

    # rank 
    d["rank"] = range(len(d))

    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=d["rank"],
        y=d[metric],
        mode='lines+markers',
        line=dict(color="#5DADE2", width=3),
        marker=dict(size=7, color="#5DADE2", line=dict(color="black", width=1)),
        hovertemplate="<b>%{text}</b><br>Value: %{y:.2f}<extra></extra>",
        text=d["country"],
        showlegend=False
    ))

    if len(d) > 0:
        fig.add_annotation(
            x=d["rank"].iloc[0],
            y=d[metric].iloc[0],
            text=d["country"].iloc[0],
            showarrow=False,
            yshift=10
        )

        fig.add_annotation(
            x=d["rank"].iloc[-1],
            y=d[metric].iloc[-1],
            text=d["country"].iloc[-1],
            showarrow=False,
            yshift=10
        )

    # ======================
    # LAYOUT
    # ======================
    fig.update_layout(
        plot_bgcolor="white",
        paper_bgcolor="white",
        title=dict(
            text=f"Trend of {metric.replace('_',' ').title()} <br> Across Top {top_n} Countries (Ranked from Low to High)",
            x=0.5
        ),
        xaxis=dict(
            title="Country Rank (Low → High)",
            showgrid=True,
            gridcolor="lightgray",
            showline=True,
            linecolor="black",
            linewidth=2,
            mirror=True
        ),
        yaxis=dict(
            title=f"{metric.replace('_',' ').title()}",
            showgrid=True,
            gridcolor="lightgray",
            showline=True,
            linecolor="black",
            linewidth=2,
            mirror=True,
            range=[0, max(d[metric].max() * 1.2, 1)]
        ),
        margin=dict(l=60, r=40, t=70, b=60)
    )

    return fig, d
````
