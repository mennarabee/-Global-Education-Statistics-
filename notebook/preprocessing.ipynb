- Remove extra spaces

## 🔍 Detect Invalid Zero Values



```python
for col in ["literacy_male", "literacy_female", "reading_prof", "math_prof"]:
    print(col)
    print((df[col] == 0).sum())
    print("------")
```

```python
import numpy as np

cols = ["literacy_male", "literacy_female", "reading_prof", "math_prof"]

for col in cols:
    df[col] = df[col].replace(0, np.nan)
```

```python
print(df[cols].isnull().sum())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
print(df[cols].isnull().sum())
```

## 🧠 Feature Engineering

We create new features to better analyze education levels across countries.

### 📊 Literacy Rate
This represents the average literacy rate between males and females.

### ⚖️ Literacy Gap
This shows the difference between male and female literacy rates, helping identify gender inequality in education.


```python
df["literacy_rate"] = (df["literacy_male"] + df["literacy_female"]) / 2
df["literacy_gap"] = df["literacy_male"] - df["literacy_female"]
```

```python
print(df[["literacy_male", "literacy_female", "literacy_rate", "literacy_gap"]].head())
```

```python
df["literacy_rate"] = df["literacy_rate"].round(2)
df["literacy_gap"] = df["literacy_gap"].round(2)
```

```python
literacy_raw = pd.read_csv(
    r"C:\Users\nosar\p)visual project\data\API_SE.ADT.LITR.ZS_DS2_en_csv_v2.csv",
    skiprows=4
)

unemp_raw = pd.read_csv(
    r"C:\Users\nosar\p)visual project\data\API_SL.UEM.TOTL.ZS_DS2_en_csv_v2.csv",
    skiprows=4
)
```

```python
literacy = literacy_raw.iloc[4:]
unemp = unemp_raw.iloc[4:]
```

```python
years = [str(y) for y in range(2015, 2021)]

literacy = literacy[["Country Name"] + years]
unemp = unemp[["Country Name"] + years]
```

```python
literacy = literacy.melt(
    id_vars=["Country Name"],
    var_name="year",
    value_name="literacy_rate"
)

unemp = unemp.melt(
    id_vars=["Country Name"],
    var_name="year",
    value_name="unemployment_rate"
)
```

```python
literacy.columns = ["country", "year", "literacy_rate"]
unemp.columns = ["country", "year", "unemployment_rate"]
```

```python
# توحيد اسم الدولة
literacy["country"] = literacy["country"].str.strip().str.lower()
unemp["country"] = unemp["country"].str.strip().str.lower()

# تحويل types
literacy["year"] = literacy["year"].astype(int)
unemp["year"] = unemp["year"].astype(int)

literacy["literacy_rate"] = pd.to_numeric(literacy["literacy_rate"], errors="coerce")
unemp["unemployment_rate"] = pd.to_numeric(unemp["unemployment_rate"], errors="coerce")
```

```python
literacy = literacy.dropna(subset=["literacy_rate"])
unemp = unemp.dropna(subset=["unemployment_rate"])
```

```python
df_final = literacy.merge(
    unemp,
    on=["country", "year"],
    how="inner"
)
```

```python
edu = pd.read_csv(r"C:\Users\nosar\p)visual project\data\cleaned_education.csv")

edu["country"] = edu["country"].str.strip().str.lower()

df_final = df_final[df_final["country"].isin(edu["country"])]
```
